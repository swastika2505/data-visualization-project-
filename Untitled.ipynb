```python
top_scores= df.groupby('player_name')['goals'].sum().sort_values(ascending=False).head(10)
top_scores.plot(kind='bar')
plt.title("Top 10 Goal Scores")
plt.ylabel("Goals")
plt.show()
```

```python
top_assists=df.groupby('player_name')['assists'].sum().sort_values(ascending=False).head(10)
top_assists.plot(kind='bar')
plt.title("Top 10 Assist Providers")
plt.ylabel("Assists")
plt.show()
```

```python
sns.countplot(x='position', data=df)
plt.title("Player Position")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['age'],bins=10)
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(df['player_rating'], bins=15)
plt.title("Player Rating distribution")
plt.show()
```

```python
numeric_df= df.select_dtypes(include='number')

plt.figure(figsize=(15,10))
sns.heatmap(numeric_df.corr(), cmap='coolwarm')
plt.title("correlation Heatmap")
plt.show()
```

```python
team_goals=df.groupby('team')['goals'].sum().sort_values(ascending=False).head(10)
team_goals.plot(kind='bar')
plt.title("Top Team by Goals")
plt.ylabel("Goals")
plt.show()
```

```python
best_players=df.groupby('player_name')['tournament_rating'].mean().sort_values(ascending=False).head(10)
best_players.plot(kind='bar')
plt.title("Top players by tournament rating")
plt.ylabel("Rating")
plt.show()
```

```python
sns.scatterplot(x='goals', y='performance_score', data=df)
plt.title("Goals vs Performance Score")
plt.show()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
plt.figure(figsize=(8,5))
sns.boxplot('Outlier Detection: Player Rating')
plt.show()
```

```python
top_scorers = df.groupby('player_name')['goals'].sum()
top_scorers = top_scorers.sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
top_scorers.plot(kind='bar')
plt.title('Top 10 Goal Scorers')
plt.xlabel('Player')
plt.ylabel('Goals')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['age'], bins=10)
plt.title('Age Distribution of Players')
plt.xlabel('Age')
plt.ylabel('Number of Players')
plt.show()
```

```python
plt.figure(figsize=(10,5))
df['position'].value_counts().plot(kind='bar')
plt.title('Player Position Distribution')
plt.xlabel('Position')
plt.ylabel('Count')
plt.show()
```
